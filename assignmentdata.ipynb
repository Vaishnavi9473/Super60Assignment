```python
#1
#Question 1. Nested if Statement
#Write a program that asks the user to input a number and prints whether the number is positive and even, positive and odd, or negative.
num=int(input("Enter a number:"))
if num>=0:
    if num%2==0:
     print("The number is positive and even")
    else:
        print("The number is positive and odd")
else:
    print("The number is negative")
```

```python
#2
#Question2. Nested Loops
#Write a program that prints a 5x5 grid of asterisks (*) using nested loops.
for i in range(5):          
    for j in range(5):     
        print("*", end=" ") 
    print()                 
```

```python
#3
#Question 3: break Statement
#Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

total = 0
while True:
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break              
    total += num           

print("The sum of all numbers is:", total)
```

```python
#4
#Question 4: continue Statement
#Write a program that prints all the numbers from 1 to 10 except 5 using a for loop and continue statement.
for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
#5
#Question 5: Fibonacci Sequence
#Write a program that prints the first n Fibonacci numbers, where n is input by the user.

n = int(input("Enter how many terms you want: "))
a, b = 0, 1
print("Fibonacci sequence:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#6
#Question 6: Creating and Accessing Lists
#Create a list of the first 20 positive integers. Print the list.
numbers = list(range(1, 21))
print("The first 20 positive integers are:")
print(numbers)
```

```python
#7
#Question 7: Accessing List Elements
#Print the first, middle, and last elements of the list created in Question 6.
first = numbers[0]
middle = numbers[len(numbers)//2 - 1 : len(numbers)//2 + 1]
last = numbers[-1]
print("First element:", first)
print("Middle element(s):", middle)
print("Last element:", last)
```

```python
#8
#Question 8: List Slicing
#Print the first five elements, the last five elements, and the elements from index 5 to 15 of the list created in Assignment 1.
print(numbers[:5])
print(numbers[-5:])
print(numbers[4:15])
```

```python
#9
#Question 9: Filtering Lists
#Create a new list containing only the even numbers from the list created in Question 6 using a list comprehension. Print the new list.
even_numbers = [x for x in numbers if x % 2 == 0]
print("Even numbers:", even_numbers)
```

```python
#10
#Question 10: List Manipulation
#Create a list of the first 10 positive integers. Remove the elements at indices 2, 4, and 6, and insert the element '99' at index 5. Print the modified list.
nums = list(range(1, 11))

# Remove elements at indices 2, 4, and 6
for index in sorted([2, 4, 6], reverse=True):
    nums.pop(index)

# Insert 99 at index 5
nums.insert(5, 99)

print("Modified list:", nums)
```

```python
#14
#Question 14: Nested Tuples
#Create a nested tuple representing a 3x3 matrix and print the matrix. Access and print the element at the second row and third column.

matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

print("Matrix:")
for row in matrix:
    print(row)

print("Element at second row, third column:", matrix[1][2])
```
